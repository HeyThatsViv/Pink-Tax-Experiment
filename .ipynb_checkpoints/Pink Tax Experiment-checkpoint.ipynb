```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
long = pd.read_csv('PinkTaxExperimentLong.csv')
```

```python
long
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax = sns.boxplot(data=long, x='Item', y='Price', hue='Gender')
```

```python
col_order = ['Canvas Sneaker', 'Denim Jeans', 'White T-Shirt', 'Underwear', 'Socks']
sns.set(font_scale=2)

g = sns.catplot(data=long, x='Item', y='Price', hue='Gender', kind='bar', 
                order=col_order, height=8, aspect=2, ci=None)
g.set_axis_labels('Item', 'Average Price')
g.fig.suptitle('Average Prices of Items By Gender', fontsize=36, y=1.05);
```
